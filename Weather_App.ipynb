```python
from tkinter import *
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk 
import requests

root = Tk()
root.title("Weather App")
root.geometry("900x700+300+200")
root.resizable(False,False)

def get_weather():

    try:
    
        city = usr_entry.get()
        api = "https://api.openweathermap.org/data/2.5/weather?q="+city+"&appid=a526d9d1e98f6136e09437c769285861" #API token süresi geçerse çalışmaz. Bu durumda, yeni API token alınması gerekir.
        
        json_data = requests.get(api).json()
        weather = json_data['weather'][0]['main']
        description = json_data['weather'][0]['description']
        temp_kelvin = json_data['main']['temp']
        temp_celcius = round(temp_kelvin -273.25,1)
        pressure = json_data['main']['pressure']
        humidity = json_data['main']['humidity']
        wind = json_data['wind']['speed']
        icon = json_data['weather'][0]['icon']
        city = json_data['name']
        country = json_data['sys']['country']
    
        t.config(text=f'{temp_celcius}°')
        c.config(text='Weather: ' + weather)
        w.config(text=wind)
        h.config(text=humidity)
        d.config(text=description)
        p.config(text=pressure)

        new_icon = PhotoImage(file=f"icons/{icon}.png")
        weather_image.config(image=new_icon, bg='#D0F0C0')
        weather_image.image = new_icon 

    except Exception as e:
        messagebox.showerror("Invalid Request",f"We cannot find a city named {city}")
    
input_text = Label(root,text='Search for a city...',fg='white',bg='gray',font=('Monaco',15,'bold'))
input_text.place(x=0,y=0)

search_template_img = PhotoImage(file="icons/search.png")
search_template_imgl = Label(image=search_template_img)
search_template_imgl.place(x=25,y=50)

usr_entry = tk.Entry(root,justify='center',width=17,fg='white',bg='black',font=("Monaco",25,"bold"))
usr_entry.place(x=50,y=70)

search_icon = PhotoImage(file="icons/search_icon.png")
searchimage = Button(image=search_icon,borderwidth=1,command=get_weather)
searchimage.place(x=400,y=62)

weather_image = Label(root)
weather_image.place(x=150, y=300)

box_image = PhotoImage(file="icons/box.png")
box_imagel = Label(image=box_image)
box_imagel.pack(side=BOTTOM)

label1 = Label(root,text="Wind",font=("Monaco",15,"bold"),bg='light blue')
label1.place(x=120,y=610)

label2 = Label(root,text="Humidity",font=("Monaco",15,"bold"),bg='light blue')
label2.place(x=250,y=610)

label3 = Label(root,text="Pressure",font=("Monaco",15,"bold"),bg='light blue')
label3.place(x=450,y=610)

label4 = Label(root,text="Description",font=("Monaco",15,"bold"),bg='light blue')
label4.place(x=630,y=610)

t = Label(root, font=("Monaco", 70, "bold"), fg="#556B2F")
t.place(x=300, y=300)

c = Label(root, font=("Monaco", 15, "bold"))
c.place(x=300, y=400)

w = Label(root, text="...", font=("Monaco", 15, "bold"), bg="#1ab5ef")
w.place(x=125, y=650)

h = Label(root, text="...", font=("Monaco", 15, "bold"), bg="#1ab5ef")
h.place(x=275, y=650)

p = Label(root, text="...", font=("Monaco", 15, "bold"), bg="#1ab5ef")
p.place(x=475, y=650)

d = Label(root, text="...", font=("Monaco", 15, "bold"), bg="#1ab5ef")
d.place(x=610, y=650)

def main():
    root.mainloop()
main()
```
